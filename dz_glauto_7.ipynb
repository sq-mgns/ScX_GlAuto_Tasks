```python
class CanSwim:
    def swim(self):
        print(f"{self.name} умеет плавать!")

class CanDance:
    def dance(self):
        print(f"{self.name} умеет танцевать!")

class CanDie:
    def die(self):
        print(f"{self.name} умер.")

class CanCan:
    def can(self):
        print(f"{self.name} может что-то делать!")

class CanFlight:
    def flight(self):
        print(f"{self.name} умеет летать!")

class CanNotTalk:
    def nottalk(self):
        print(f"{self.name} не может говорить.")

class CanTalk:
    def talk(self):
        print(f"{self.name} может говорить!")
```

```python
class BaseCharacter(CanCan, CanDie):
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Привет, меня зовут - {self.name}")

class BaseInAnctionCaracter(BaseCharacter, CanFlight):
    def __init__(self, name, panache):
        super().__init__(name)
        self.panache = panache

    def greet(self):
        print(f"Привет, у меня есть - {self.panache}")

class BaseFunkoPop(BaseCharacter, CanNotTalk):
    def __init__(self, name, material):
        super().__init__(name)
        self.material = material

    def greet(self):
        print(f"Привет, мой материал - {self.material}")

class BaseCosplayer(BaseCharacter, CanSwim):
    def __init__(self, name, soul):
        super().__init__(name)
        self.soul = soul

    def greet(self):
        print(f"Привет, моя душа - {self.soul}")

class BaseHuman(BaseCharacter, CanTalk):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age

    def greet(self):
        print(f"Привет, мой возраст - {self.age}")
```

```python
class Shrek(BaseInAnctionCaracter, CanTalk, CanSwim):
    def __init__(self):
        super().__init__("Шрэк", "сила огра")

    def greet(self):
        print("я Шрэк")

class PussInBoots(BaseInAnctionCaracter, CanTalk):
    def __init__(self):
        super().__init__("Кот в Сапогах", "шпага")

    def greet(self):
        print("я Кот в Сапогах")

class Donkey(BaseCharacter, CanTalk, CanDance):
    def __init__(self):
        super().__init__("Ослик")

    def greet(self):
        print("я Ослик")

class JackHorner(BaseHuman, CanFlight):
    def __init__(self):
        super().__init__("Джек Хорни", "96 лет")

    def greet(self):
        print("я Джек Хорни")

class ShrekFunkoPop(BaseFunkoPop):
    def __init__(self):
        super().__init__("Шрэк", "зеленый бетон")

class PussInBootsFunkoPop(BaseFunkoPop):
    def __init__(self):
        super().__init__("Кот в Сапогах", "оранжевый азбест")

class DonkeyFunkoPop(BaseFunkoPop):
    def __init__(self):
        super().__init__("Ослик", "серый гудрон")

class JackHornerFunkoPop(BaseFunkoPop):
    def __init__(self):
        super().__init__("Джек Хорни", "фиол. резина")

class ShrekCosplayer(BaseCosplayer):
    def __init__(self):
        super().__init__("Шрэк", "самбади ванс толд ми")

class PussInBootsCosplayer(BaseCosplayer):
    def __init__(self):
        super().__init__("Кот в Сапогах", "6 жизней")

class DonkeyCosplayer(BaseCosplayer):
    def __init__(self):
        super().__init__("Ослик", "пусто, как и в голове")

class JackHornerCosplayer(BaseCosplayer):
    def __init__(self):
        super().__init__("Джек Хорни", "злая душа")
```

```python
shrek = Shrek()
shrek_pop = ShrekFunkoPop()
shrek_cos = ShrekCosplayer()

puss = PussInBoots()
puss_pop = PussInBootsFunkoPop()
puss_cos = PussInBootsCosplayer()

donkey = Donkey()
donkey_pop = DonkeyFunkoPop()
donkey_cos = DonkeyCosplayer()

jack = JackHorner()
jack_pop = JackHornerFunkoPop()
jack_cos = JackHornerCosplayer()
```
